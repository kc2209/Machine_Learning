Here we will be predicting the salary based on years of experience using simple linear regression.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Salary_Data.csv")
```

Dividing the dataset into dependent and independent variables

```python
x = dataset.iloc[:,:-1].values # selecting years of experience (independent variable)
y = dataset.iloc[:,-1].values  # selecting salary (dependent variable)
```

Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

Selecting the model and training it.

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  # assigning the function to a variable
regressor.fit(x_train,y_train)  # using fit function we can make our model learn from the learning samples.
```

Evaluating the model using testing data. Predicting the test set

```python
y_pred = regressor.predict(x_test)
```

Visualizing the results. Both training set and testing set

```python
plt.scatter(x_train,y_train,color='red')
#plt.scatter(x_test,y_test,color='green')
plt.plot(x_train,regressor.predict(x_train),color = 'blue')
plt.title("Salary vs Experience(Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,regressor.predict(x_train),color = 'blue')   # same prediction line for comparison
plt.title("Salary vs Experience(Test set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
